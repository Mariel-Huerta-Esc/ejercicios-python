```python
"""
EJERCICIO 4

    Guillermo tiene N dólares. (Es decir, el usuario tiene que poner la cantidad
que tiene Guillermo), Luis tiene la mitad de lo que posee Guillermo. Juan tiene
la mitad de lo que poseen Luis y Guillermo juntos. Hacer un programa que
calcule e imprima la cantidad de dinero que tienen entre los tres.
                                                                             """


guillermo = int(input("Hola, dime cuánto dinero tiene Guillermo para poder decirte cuánto dinero tienen los demás "))

#haciendo las operaciones
luis = (guillermo / 2)
juan = ((luis + guillermo)/2)


# imprimiendo cuánto tiene cada quién:
print(f"el dinero que tiene Guillermo: {guillermo}")
print(f"el dinero que tiene Luis: {luis}")
print(f"el dinero que tiene Juan: {juan}")
```

```python
"""
EJERCICIO 5

La calificación final de un estudiante de informática se determina en función
de sus calificaciones en cuatro aspectos de su rendimiento académico: partici-
pación, primer examen parcial, segundo examen parcial y examen final. Las
ponderaciones para cada aspecto son las siguientes: Participación: 10 % , primer
examen parcial: 25 %, segundo examen parcial: 25 %, examen final: 40 %.
Desarrolla un programa que solicite al usuario ingresar las calificaciones obtenidas en
cada uno de estos aspectos. El programa debe calcular e imprimir la calificación
final del estudiante basándose en las ponderaciones mencionadas.
1
                              """
participacion = int (input("Ingresa tu calificacion del 1 al 10 en participacion:  "))
examen1 = int(input("Ingresa tu calificacion del 1 al 10 del examen 1:  "))
examen2 = int(input("Ingresa tu calificacion del 1 al 10 de examen 2:  "))
examenfinal = int(input("Ingresa tu calificacion del 1 al 10 de tu examen final:  "))

# haciendo las operaciones:
resultado1 = participacion * 0.1
resultado2 = examen1 * 0.25
resultado3 = examen2 * 0.25
resultado4 = examenfinal * 0.40


print("porcentaje de participación:  ", resultado1)
print("porcentaje de examen1:  ", resultado2)
print("porcentaje de examen2:  ", resultado3)
print("porcentaje de examen final:  ", resultado4)
```

```python
"""
EJERCICIO 8

Construye un programa que tome como entrada un número total de horas
y calcule la cantidad equivalente en semanas, días y horas. Por ejemplo, si se
ingresan 1000 horas, el programa debe mostrar que esto equivale a 5 semanas,
6 días y 16 horas
"""


# variable
horas = int(input("Ingresa las horas que deseas convertir:   "))
semana = 168
dia = 24

# calculando semanas, dias y horas, poniendo // se usa division entera para que
# no haya decimales

semanas = horas // semana
resto = horas % semana
dias =  resto // dia
horas = resto % dia

print("esta es la conversion:  ")
print("en semanas:  ", semanas)
print("en dias:  ", dias)
print("en horas:  ", horas)

```

```python
"""
EJERCICIO 9

Escribir un programa que lea un entero positivo n, introducido por el usuario
y después muestre en pantalla la suma de todos los enteros desde 1 hasta n. La
suma de los n primeros enteros positivos puede ser calculada de la siguiente
forma: suma = n(n+1)/2

"""


# variable
n = int(input("Ingresa un número:  "))

#operacion
numerador = n * (n+1)
denominador = 2
total = numerador / denominador

print("El resultado es:  ", total)

```
